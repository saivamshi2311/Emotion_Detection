```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
model = load_model('emotion_model_MobileNetV2.h5')
emotion_labels = ['Angry', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(1)
```

```python
print("Starting real-time emotion recognition. Press 'q' to exit.")
```

```python
while True:

    ret, frame = cap.read()
    if not ret:
        print("Failed to capture frame. Exiting...")
        break

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        roi_gray = gray_frame[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]
        roi_resized = cv2.resize(roi_gray, (128, 128))
        roi_normalized = roi_resized / 255.0
        roi_input = np.expand_dims(roi_normalized, axis=(0, -1))
        predictions = model.predict(roi_input)
        emotion_index = np.argmax(predictions)
        emotion_label = emotion_labels[emotion_index]
        confidence = np.max(predictions) * 100
        label_text = f"{emotion_label}: {confidence:.2f}%"

        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2) 
        cv2.putText(frame, label_text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)  

    cv2.imshow('Real-Time Facial Emotion Recognition', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
cap.release()
cv2.destroyAllWindows()
```
